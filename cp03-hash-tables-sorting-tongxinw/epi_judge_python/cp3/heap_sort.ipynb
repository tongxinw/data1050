1. The original BinHeap class is not a in-place modification class, as it adds an element 0 in the front and generates a new array to modify. Also, the index of the orignal BinHeap class needs to be changed as the 0 matters. 

2. The time complexity for building a heap is O(N) and the time complexity for sort it is O(logN), thus the time complexity for heap_sort is O(NlogN). Since heap_sort is inplace sort, the space complexity is O(1). The time complexity of heap_sort is the same for best and the worst cases, and both of them are O(NlogN). The best case would be the input is a ascending list, and the worst case would be having a descending input. 

```python
import perfplot # pip install perfplot --user
from myBinHeap import myBinHeap
from heap_sort import heap_sort

percDown_no_key = myBinHeap.percDown_without_key # save default value

def heap_sort_argmin(L):
    myBinHeap.percDown_without_key = myBinHeap.percDown_argmin_without_key
    R = heap_sort(L[:])
    myBinHeap.percDown_without_key = percDown_no_key
    return(R)

def heap_sort_minChild(L):
    myBinHeap.percDown_without_key=myBinHeap.percDown_minChild_without_key
    R = heap_sort(L[:])
    myBinHeap.percDown_without_key = percDown_no_key
    return(R)

def heap_sort_no_key(L): # uses the default percDown_without_key
    R = heap_sort(L[:])
    return(R)

def heap_sort_identity_key(L): # uses percDown_with_key
    R = heap_sort(L[:], key=lambda x:x)
    return(R)

b = perfplot.bench(
    setup=lambda n: list(reversed(range(n))),
    kernels=[
        heap_sort_no_key,
        heap_sort_argmin,
        heap_sort_minChild,
        heap_sort_identity_key
        ],
    n_range=[10**k for k in range(1,5)],
#     logx=False,
#     logy=False,
    title = 'heap_sort argmin vs heap_sort minChild'
    )

b.show()
```
